```python
%pip install openpyxl
import pandas as pd
import os
import shutil

# Path to the dataset folder containing the images
dataset_folder = 'C:\\Fetal_Brain\\FetalImages'

# Path to the Excel file
excel_file = 'C:\\Fetal_Brain\\FETAL_PLANES_DB_data.xlsx'

# Read the Excel file into a DataFrame
df = pd.read_excel(excel_file)

# Create the three separate folders for each plane
output_folder_plane1 = 'C:\\Fetal_Brain\\CereOriginal'
output_folder_plane2 = 'C:\\Fetal_Brain\\ThalamusOriginal'
output_folder_plane3 = 'C:\\Fetal_Brain\\VentriOriginal'

# Create the output folders if they don't exist
os.makedirs(output_folder_plane1, exist_ok=True)
os.makedirs(output_folder_plane2, exist_ok=True)
os.makedirs(output_folder_plane3, exist_ok=True)

# Iterate over the rows in the DataFrame
for index, row in df.iterrows():
    image_name = row['Image_name']
    plane_name = row['Brain_plane']

    # Determine the destination folder based on the plane name
    if plane_name == 'Trans-cerebellum':
        destination_folder = output_folder_plane1
    elif plane_name == 'Trans-thalamic':
        destination_folder = output_folder_plane2
    elif plane_name == 'Trans-ventricular':
        destination_folder = output_folder_plane3
    else:
        # Skip the image if the plane name is not recognized
        continue

    # Construct the source and destination paths
    source_path = os.path.join(dataset_folder, image_name+'.png')
    destination_path = os.path.join(destination_folder, image_name+'.png')

    # Copy the image to the destination folder
    shutil.copy2(source_path, destination_path)

print('Images segregation complete.')
```

```python
import pandas as pd
import os
from PIL import Image

# Read the CSV file
df = pd.read_csv('C:\\Fetal_Brain\\CropData.csv')

# Input and output folders
input_folder = 'C:\\Fetal_Brain\\VentriOriginal'
output_folder = 'C:\\Fetal_Brain\\CroppedImages'

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    # Extract filename and region shape attributes
    filename = row['filename']
    region_shape_attributes = eval(row['region_shape_attributes'])
    
    # Open the image
    image_path = os.path.join(input_folder, filename)
    image = Image.open(image_path)
    
    # Extract region shape attributes
    x = region_shape_attributes['x']
    y = region_shape_attributes['y']
    width = region_shape_attributes['width']
    height = region_shape_attributes['height']
    
    # Crop the image
    cropped_image = image.crop((x, y, x + width, y + height))
    
    # Save the cropped image
    output_path = os.path.join(output_folder, filename)
    cropped_image.save(output_path)

    # Close the image
    image.close()
```

```python
import pandas as pd
import os
from PIL import Image
import cv2
import numpy as np

# Input and output folders
input_folder = 'C:\\Fetal_Brain\\VentriOriginal'
output_folder = 'C:\\Fetal_Brain\\EdgedImages'

# Read the dataframe
df = pd.read_csv('C:\\Fetal_Brain\\CropData.csv') 

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    # Extract filename and region shape attributes
    filename = row['filename']
    region_shape_attributes = eval(row['region_shape_attributes'])
    
    # Open the image
    image_path = os.path.join(input_folder, filename)
    image = Image.open(image_path)
    
    # Extract region coordinates
    x = region_shape_attributes['x']
    y = region_shape_attributes['y']
    width = region_shape_attributes['width']
    height = region_shape_attributes['height']
    
    # Crop the image
    edgedImage = image.crop((x, y, x + width, y + height))

    # Convert PIL image to numpy array
    img_np = np.array(edgedImage)

    # Convert image to grayscale
    img_gray = cv2.cvtColor(img_np, cv2.COLOR_BGR2GRAY)

    # Apply Sobel edge detection
    gradx = cv2.Sobel(img_gray, -1, 1, 0)
    grady = cv2.Sobel(img_gray, -1, 0, 1)
    gradxy = cv2.addWeighted(gradx, 0.4, grady, 0.2, 0)

    # Write the edge-detected image
    edge_output_path = os.path.join(output_folder, 'sobel_' + filename)  # Define output path for edge-detected image
    cv2.imwrite(edge_output_path, gradxy)
    
    # Save the cropped image
    cropped_output_path = os.path.join(output_folder, filename)
    edgedImage.save(cropped_output_path)

    # Close the image
    image.close()
```
